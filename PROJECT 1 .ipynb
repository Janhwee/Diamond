```python
def IQR(x):
    q1=x.quantile(.25)
    q3=x.quantile(.75)
    IQR=q3-q1
    lf=q1-(1.5*IQR)
    uf=q3+(1.5*IQR)
    print("lf-->",lf,"uf-->",uf)
```

```python
df1["price_log"]=np.log(df1["price"])
```

```python
y = df1["price_log"] # target variable/ dependent variables 
x = df1[['carat', 'depth', 'table', 'x', 'y',
       'z' ]] # independent variable
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=80)
```

```python
x_train=sm.add_constant(x_train)
x_test=sm.add_constant(x_test)
model2=sm.OLS(y_train,x_train).fit()
```

```python
print(model2.summary())
```

```python
actual=np.exp(y_train)
pre=np.exp(model2.predict(x_train))
error=actual-pre
```

```python
print("MSE",np.mean(np.square(error)))
print("RMSE",np.sqrt(np.mean(np.square(error))))
print("MAE",np.mean(np.abs(error)))
```
